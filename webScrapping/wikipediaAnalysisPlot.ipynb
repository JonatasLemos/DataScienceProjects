# Plots Citizens

```python
import pandas as pd
import matplotlib.pyplot as plt
```

## Reading CSV and sorting

```python
cities_with_famous = pd.read_csv("merged_cities_famous.csv")
cities_with_famous = cities_with_famous.drop(["Unnamed: 0"],axis=1)
cities_with_famous = cities_with_famous.sort_values(by="PESSOAS NA WIKIPEDIA", ascending=True)
```

## Ploting 

```python
plt.barh(cities_with_famous["NAME"].tail(10)
         ,cities_with_famous["PESSOAS NA WIKIPEDIA"].tail(10))
plt.title("Pessoas com página na Wikipedia - Minas Gerais")
plt.xlabel('Número de pessoas')
plt.ylabel('Cidade')
plt.grid(lw="0.2")
plt.show()
```

## Filtering to calculate rate per 100000

```python
cities_with_famous = cities_with_famous[cities_with_famous['PESSOAS NA WIKIPEDIA'] > 0]
cities_with_famous["PESSOAS NA WIKIPEDIA POR 100000"] = (cities_with_famous["PESSOAS NA WIKIPEDIA"] / \
                                                  cities_with_famous["POPULAÇÃO ESTIMADA"])*100000
cities_with_famous = cities_with_famous.sort_values(by="PESSOAS NA WIKIPEDIA POR 100000", ascending=True)
```

## Ploting rate

```python
plt.barh(cities_with_famous["NAME"].tail(10)
         ,cities_with_famous["PESSOAS NA WIKIPEDIA POR 100000"].tail(10))
plt.title("Maiores taxas de pessoas na Wikipedia por 100000 habitantes - MG")
plt.xlabel('Taxa por 100000 habitantes')
plt.ylabel('Cidade')
plt.grid(lw="0.2")
plt.show()
```

```python
cities_with_famous = cities_with_famous.sort_values(by="PESSOAS NA WIKIPEDIA POR 100000", ascending=False)
plt.barh(cities_with_famous["NAME"].tail(10)
         ,cities_with_famous["PESSOAS NA WIKIPEDIA POR 100000"].tail(10))
plt.title("Menores taxas de pessoas na Wikipedia por 100000 habitantes - MG")
plt.xlabel('Taxa por 100000 habitantes')
plt.ylabel('Cidade')
plt.grid(lw="0.2")
plt.show()
```

```python
cities_with_famous = cities_with_famous[cities_with_famous["POPULAÇÃO ESTIMADA"] > 50000]
cities_with_famous = cities_with_famous.sort_values(by="PESSOAS NA WIKIPEDIA POR 100000", ascending=True)
plt.barh(cities_with_famous["NAME"].tail(10)
         ,cities_with_famous["PESSOAS NA WIKIPEDIA POR 100000"].tail(10))
plt.suptitle("Maiores taxas de pessoas na Wikipedia por 100000 habitantes - MG")
plt.title("Cidades com mais de 50000 habitantes")
plt.xlabel('Taxa de pessoas')
plt.ylabel('Cidade')
plt.grid(lw="0.2")
plt.show()
```

```python
cities_with_famous = cities_with_famous.sort_values(by="PESSOAS NA WIKIPEDIA POR 100000", ascending=False)
plt.barh(cities_with_famous["NAME"].tail(10)
         ,cities_with_famous["PESSOAS NA WIKIPEDIA POR 100000"].tail(10))
plt.suptitle("Menores taxas de pessoas na Wikipedia por 100000 habitantes - MG")
plt.title("Cidades com mais de 50000 habitantes")
plt.xlabel('Taxa de pessoas')
plt.ylabel('Cidade')
plt.grid(lw="0.2")
plt.show()
```
